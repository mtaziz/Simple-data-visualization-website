```python
zone_list = []
for i in position.find():
    zone_list.append(i['zone'])
zone_index=list(set(zone_list))
print str(zone_index).replace('u\'','\'').decode('unicode-escape')
    
```

```python
post_times =[]
for index in zone_index:
    post_times.append(zone_list.count(index))
print post_times
```

```python
def data_gen(types):
    length = 0
    if length<= len(zone_index):
        for zone,times in zip(zone_index, post_times):
            data= {
                'name': zone,
                'data': [times],
                'type': types,
            }

            yield data
            length += 1
```

```python
data_gen('column')
```
