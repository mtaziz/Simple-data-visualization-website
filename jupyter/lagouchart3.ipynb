```python
import pymongo
import charts
```

```python
client = pymongo.MongoClient('localhost',27017)
lagou = client['LaGou']
position = lagou['positionY']
```

```python
for i in position.find():
    print str(i).replace('u\'','\'').decode('unicode-escape')
```

```python
pipeline = [
    {'$match':{'$and':[{'createtime':'2018-07-13'},{'workyear':u'1-3年'}]}},
    {'$group':{'_id':'$salary','count':{'$sum':1}}},
    {'$sort':{'count':-1}},
    {'$limit':10}
]
# {'createtime':'2018-07-13'}
```

```python
for i in position.aggregate(pipeline):
    print str(i).replace('u\'','\'').decode('unicode-escape')
```

```python
for i in position.find({},{'_id':0,'labels':1}):
    print str(i).replace('u\'','\'').decode('unicode-escape')
```

```python
for i in position.aggregate(pipeline):
    print str(i).replace('u\'','\'').decode('unicode-escape')
```

```python
def salary_gen():
    pipeline = [
        {'$group':{'_id':'$salary','counts':{'$sum':1}}},
        {'$sort':{'counts':-1}},
        {'$limit':20}
    ]
    for i in position.aggregate(pipeline):
        yield [i['_id'],i['counts']]   
```

```python
for i in salary_gen():
    print i
```

```python
options = {
    'chart': {'zoomType':'xy'},
    'title': {'text':'LaGou深圳市python招聘各地区一个月发布职位薪资情况'},
    'subtitle': {'text':'可视化图表'}
    }
series = [
    {
        'type': 'pie',
        'name': '薪资情况',
        'data': [i for i in salary_gen()]
        
        
    }
        
]
charts.plot(series, options=options,show='inline')
```

```python
def workyear_gen():
    pipeline = [
        {'$group':{'_id':'$workyear','counts':{'$sum':1}}},
        {'$sort':{'counts':-1}},
#         {'$limit':20}
    ]
    for i in position.aggregate(pipeline):
        yield [i['_id'],i['counts']]   
```

```python
for i in workyear_gen():
    print str(i).replace('u\'','\'').decode('unicode-escape')
```

```python
options = {
    'chart': {'zoomType':'xy'},
    'title': {'text':'LaGou深圳市python招聘各地区一个月发布职位工作年限要求'},
    'subtitle': {'text':'可视化图表'}
    }
series = [
    {
        'type': 'pie',
        'name': '年限要求',
        'data': [i for i in workyear_gen()]
        
        
    }
        
]
charts.plot(series, options=options,show='inline')
```
